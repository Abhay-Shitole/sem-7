```python
# Function and its derivative
def f(x):
    return (x + 3)**2

def df(x):
    return 2 * (x + 3)

# Initial guess
x = 2

# Learning rate
alpha = 0.1

# Number of iterations
epochs = 50
x_history = [2]

print("Iteration\tX Value\t\tFunction Value")
for i in range(epochs):
    grad = df(x)
    x = x - alpha * grad
    x_history.append(x)
    print(f"{i+1}\t\t{x:.6f}\t\t{f(x):.6f}")

print("\nLocal minimum occurs at x =", round(x, 4))


import matplotlib.pyplot as plt

x_vals = [i for i in range(-10,5)]
y_vals = [f(i) for i in x_vals]

plt.plot(x_vals , y_vals, label="y=(x+3)^2")
plt.scatter(x_history , [f(i) for i in x_history] , color="red" , label="convergence path")
plt.xlabel("x")
plt.ylabel("y")
plt.title("gradient descent convergences")
plt.legend()
plt.show()
```
