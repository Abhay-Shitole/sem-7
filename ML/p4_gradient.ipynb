```python
# Gradient Descent to find local minima of y = (x + 3)^2

# Import required library
import matplotlib.pyplot as plt

# Define function and its derivative
def f(x):
    return (x + 3)**2

def df(x):
    return 2 * (x + 3)

# Initialize parameters
x = 2          # starting point
learning_rate = 0.1
iterations = 25

# To store path of x values
x_history = []

# Gradient Descent loop
for i in range(iterations):
    grad = df(x)
    x = x - learning_rate * grad
    x_history.append(x)
    print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")

# Final output
print("\nLocal minima occurs at x =", round(x, 4))
print("Minimum value of function y =", round(f(x), 4))

# Plot the function and convergence path
x_vals = [i for i in range(-10, 5)]
y_vals = [f(i) for i in x_vals]

plt.plot(x_vals, y_vals, label='y = (x+3)^2')
plt.scatter(x_history, [f(i) for i in x_history], color='red', label='Convergence path')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Convergence')
plt.legend()
plt.show()
```
